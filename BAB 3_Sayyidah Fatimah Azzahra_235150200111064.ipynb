### Soal 1
**1. Download Glass dataset dari https://archive.ics.uci.edu/dataset/42/glasstidentification**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/glass/glass.data"
column_names = ["Id", "RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe", "Type"]
```

#### Penjelasan

Di sini, saya mengunduh dataset dari url yang diberikan.

### Soal 2

**2. Masukkan data Glass ke dataframe menggunakan Pandas.**

```python
data = pd.read_csv(url, names=column_names)
```

#### Penjelasan

Saya menjadikan data yang diunduh tadi dalam sebuah dataframe dengan nama kolom yang saya tentukan juga menggunakan Pandas.

### Soal 3

**3. Tampilkan 10 data pertama**

```python
data.head(10)
```

#### Penjelasan

Saya menampilkan 10 data pertama menggunakan fungsi head() pada dataframe data.

### Soal 4

**4. Hitung rata-rata nilai per variabel untuk setiap data**

```python
# Jawaban Soal 4

rata_rata = data.mean()
print(rata_rata)
```

#### Penjelasan

Saya menghitung rata-rata nilai per variabel menggunakan fungsi mean() pada dataframe data.

### Soal 5

**5. Hitung rata-rata nilai per variabel untuk setiap data dikelompokkan berdasarkan Type**

```python
# Jawaban Soal 5
rata_rata_per_type = data.groupby('Type').mean()

print(rata_rata_per_type)
```

#### Penjelasan

Saya menghitung rata-rata nilai per variabel untuk setiap data dikelompokka berdasarkan type menggunakan fungsi groupby('Type'), yang kemudian dirata-ratakan menggunakan fungsi mean() dan disimpan dalam variabel rata_rata_per_type.

### Soal 6

**6. Buatlah plot bertipe 'line' untuk masing-masing variabelnya. Referensi: Pandas Plot**

```python
# Jawaban Soal 6

fig, axes = plt.subplots(nrows=len(data.columns), ncols=1, figsize=(12, 4*len(data.columns)))

for i, column in enumerate(data.columns):
    data[column].plot(kind='line', ax=axes[i])
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

Jadi, saya membuat beberapa subplot untuk setiap kolom dalam dataframe 'data'. Setiap subplot menampilkan grafik garis untuk satu kolom. Sumbu x diberi label sesuai nama kolom, sumbu y diberi label 'Count'. Ukuran gambar disesuaikan dengan jumlah kolom. Kemudian, subplot ditampilkan menggunakan metode show() dan tight_layout() untuk mengatur tata letak yang rapi.
